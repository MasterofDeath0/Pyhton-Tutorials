## NumPy
NumPy stands for Numerical Python & is the core library for nummeric & scientific computing
It consists of Multidimensional Array objects & a colllection of routines for processing those arrays

```python
import numpy as np
```

# Single Dimensional Array

```python
n1 = np.array([10,20,30,40])
n1
```

# Multi Dimensional Array

```python
n2 = np.array([[10,20,30,40],[50,60,70,80]])
n2
```

# Array with Element Zero and Same Number


```python
# array with all elements zero
n3 = np.zeros((3,4))
n3
```

```python
# array with all elements with same number
n4 = np.full((3,4),5)
n4
```

# Array with elements in a range of number

```python
n5 = np.arange(10,20,2)
n5
```

# Array with random numbers

```python
n6 = np.random.randint(1,100,10)
n6
```

```python
n7 = np.random.randint(1,10,size=(3,4))
n7
```

# Joining Numpy Arrays

```python
## Vertical Stack
n1 = np.array([10,20,30,40,50])
n2 = np.array([1,2,3,4,5])
np.vstack((n2,n1))
```

```python
## Horizontal Stack
np.hstack((n1,n2))
```

```python
## Column Stack
np.column_stack((n1,n2))
```

# Intersection & Difference


```python
n1 = np.array([10,20,30,40,50])
n2 = np.array([40,50,60,70,20])
np.intersect1d(n1,n2)
```

```python
np.union1d(n1,n2)
```

```python
np.setdiff1d(n1,n2)
```

```python
np.setdiff1d(n2,n1)
```

# Sum of Two Arrays


```python
n1=np.array([10,20])
n2=np.array([30,40])
np.sum([n1,n2])
```

```python
np.sum([n1,n2],axis=0)
# Column Sum
```

```python
np.sum([n1,n2],axis=1)
# Row Sum
```

```python
np.diff([n1,n2],axis=0)
```

```python
np.diff([n1,n2],axis=1)
```

# Mathematical Operations

```python
n1=np.array([10,20])
n1=n1+1
n1
```

```python
n1=np.array([10,20])
n1=n1-1
n1
```

```python
n1=np.array([10,20])
n1=n1*2
n1
```

```python
n1=np.array([10,20])
n1=n1//2
n1
```

```python
n1=np.array([10,20,30,40,50])
np.mean(n1)
```

```python
np.median(n1)
```

```python
np.std(n1)
```
